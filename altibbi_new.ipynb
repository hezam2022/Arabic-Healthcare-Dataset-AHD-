```python

#https://altibbi.com/%D8%A7%D8%B3%D8%A6%D9%84%D8%A9-%D8%B7%D8%A8%D9%8A%D8%A9?page=
start=108000
end=115500
f=start
s=start

upperframe=[]

for page in range(start, end):
    print('processing page:', page)
    url ="https://altibbi.com/%D8%A7%D8%B3%D8%A6%D9%84%D8%A9-%D8%B7%D8%A8%D9%8A%D8%A9?page="+str(page)

    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    frame6=[]
    links=soup.find_all('div',attrs={'class':'question-item-container'})
    #print(links)

    for j in links:
      link=j.find('a',attrs={'class':'question-text mt-17 id-5-5'})['href'].strip()
      #print(Link)
      question=j.find('h2').text.strip()
      category=j.find('a',attrs={'class':'category-info d-block'}).text.strip()
      questionerAge=j.find('div',attrs={'class':'age'}).text.strip()
      doctorName=j.find('a',attrs={'class':'d-block dr-name-fqaa'}).text.strip()
      view=j.find('p',attrs={'class':'ml-auto views-count'}).text.strip()
      date=j.find('time').text.strip()
      doctorAnswer=j.find('div',attrs={'class':'hide'}).text.strip()
      frame6.append((question,doctorAnswer,category,questionerAge,doctorName,view,date,link))

    upperframe.extend(frame6)
    if (f%2000)==0:
      data=pd.DataFrame(upperframe, columns=['Question','Doctor-answer','Category','Questioner_Age','Doctor_Name','View','Date','Link'])
      data.to_csv('/content/drive/MyDrive/altibbi/new/altibbi_from_'+str(s)+"_to_"+str(f)+'.csv',encoding='utf-8-sig',index=False)
      s=f
      upperframe=[]
    f=f+1

data=pd.DataFrame(upperframe, columns=['Question','Doctor-answer','Category','Questioner_Age','Doctor_Name','View','Date','Link'])
data.to_csv('/content/drive/MyDrive/altibbi/new/altibbi_from_'+str(s)+"_to_"+str(f)+'.csv',encoding='utf-8-sig',index=False)
data.head()
```
